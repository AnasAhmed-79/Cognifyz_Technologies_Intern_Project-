# Task: Feature Engineering

## Extract additional features from the existing columns, such as the length of the restaurant name or address.

```python
# Add a new column 'Restaurant Name Length' to store the length of restaurant names
df['Restaurant Name Length'] = df['Restaurant Name'].apply(len)

# Add a new column 'Address Length' to store the length of addresses
df['Address Length'] = df['Address'].apply(len)

# Now you have two additional features: 'Restaurant Name Length' and 'Address Length'

# You can access these features in the 'data' DataFrame
print(df[['Restaurant Name', 'Restaurant Name Length', 'Address', 'Address Length']])

# Save the modified DataFrame with the additional features to a new CSV file
df.to_csv('restaurants_data_with_features.csv', index=False)
```

## Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables.

```python
# Perform one-hot encoding for 'Has Table booking' and 'Has Online delivery'
data = pd.get_dummies(data, columns=['Has Table booking', 'Has Online delivery'], drop_first=True)

# The resulting DataFrame will have new columns 'Has Table booking_Yes' and 'Has Online delivery_Yes'

# You can access these new binary features in the DataFrame
print(data[['Has Table booking_Yes', 'Has Online delivery_Yes']])

# Save the modified DataFrame to a new CSV file
data.to_csv('restaurants_data_encoded.csv', index=False)
```
